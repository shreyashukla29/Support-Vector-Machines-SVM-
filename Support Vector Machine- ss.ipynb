```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop(['mean radius','mean area','mean concavity','radius error','area error','worst radius','worst area'],axis=1)
print(df.shape)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True)
```

```python
sns.countplot(df['cancer'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,df['cancer'],test_size=0.3)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()

```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```
